```python
import numpy as np
from matplotlib.pylab import *
```

$z_{n+1} = z_n^2 + c$. If $|z_{n+1}| > 2$, then the point in the complex plane at point c does not belong to the Mandelbrot set.

```python
N =500
x = np.linspace(-2, 2, N)
y = np.linspace(-2, 2,N)
```

```python
def mandelbrot(c, max_iter):
    z = 0
    for i in range(0,max_iter):
        z = z**2 + c
        if np.abs(z) > 2:
            return i
    return max_iter
```

```python
def mandel_set(x, y, max_iter):
    x_in = []
    y_in = []
    in_color = []
    x_out = []
    y_out = []
    out_color = []
    for i in range(N):
        for j in range(N):
            c = x[i] + 1j*y[j]
            itera = mandelbrot(c, max_iter)
            if itera == max_iter :
                x_in.append(x[i])
                y_in.append(y[j])
                in_color.append(max_iter)
            else:
                x_out.append(x[i])
                y_out.append(y[j])
                out_color.append(itera)
                
    return x_in, y_in, x_out, y_out, in_color, out_color
```

```python
x_in, y_in, x_out, y_out, in_color, out_color = mandel_set(x,y,10)
```

```python
figure(figsize=(10,7))
scatter(x_in, y_in, c = 'k', s = 1)
scatter(x_out, y_out, c = 'w', s = 1)
xticks(fontsize=15)
yticks(fontsize=15)
xlabel('$x$', fontsize= 20)
ylabel('$y$', fontsize= 20)
title('Mandelbrot Fractal', fontsize=20)
show()
```

Plotting the Mandelbrot fractal with number of iteration. If $|z| < 2$, return the max iteration.   

```python
figure(figsize=(10,8))
scatter(x_in, y_in, c = in_color, s = 2, cmap= 'rainbow')
scatter(x_out, y_out, c = out_color, s = 2, cmap ='rainbow')
xticks(fontsize=15)
yticks(fontsize=15)
xlabel('$x$', fontsize= 20)
ylabel('$y$', fontsize= 20)
xlim([-2,2])
ylim([-2,2])
title('Mandelbrot Fractal', fontsize=20)
show()
```

Now let's double the number of iterations to see how the fractal changes.
